# Create Line Charts

```python
# Set the seed

np.random.seed(4)
```

```python
# Create a subset of data (random samples) with the 70/30 slipt

dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# Create a new list called dev

dev
```

```python
np.random.rand(10)
```

```python
# Store 70% of the sample in the dataframe big

big = ords_prods_merge[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = ords_prods_merge[~dev]
```

```python
# Check output

len(ords_prods_merge)
```

```python
# Check if it is the same result

len(big)+ len(small)
```

```python
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Create line chart

line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```

```python
# Export line chart

line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_chart_prices.png'))
```
